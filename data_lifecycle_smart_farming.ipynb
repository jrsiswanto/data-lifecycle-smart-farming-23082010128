```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. LOAD DATA
df = pd.read_csv('plant_vase1(2).csv')

# 2. CLEANING
df = df.dropna()

# 3. VISUALISASI
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
plt.subplots_adjust(hspace=0.4)

# --- VISUALISASI 1: Heatmap Korelasi antar Sensor Moisture ---
moisture_cols = ['moisture0', 'moisture1', 'moisture2', 'moisture3', 'moisture4']
sns.heatmap(df[moisture_cols].corr(), annot=True, cmap='YlGnBu', ax=axes[0,0])
axes[0,0].set_title('1. Correlation Heatmap (Moisture Sensors)')

# --- VISUALISASI 2: Boxplot Distribusi Moisture ---
sns.boxplot(data=df[moisture_cols], ax=axes[0,1])
axes[0,1].set_title('2. Distribution of Moisture Levels')

# --- VISUALISASI 3: Line Plot Tren Moisture (100 data pertama) ---
df[moisture_cols].head(100).plot(ax=axes[1,0])
axes[1,0].set_title('3. Moisture Trend (First 100 Records)')
axes[1,0].legend(loc='upper right')

# --- VISUALISASI 4: Scatter Plot Moisture0 vs Moisture1 ---
sns.scatterplot(x='moisture0', y='moisture1', hue='irrgation', data=df, ax=axes[1,1])
axes[1,1].set_title('4. Moisture0 vs Moisture1 (by Irrigation)')

# SIMPAN HASIL
plt.savefig('moisture_summary.png')
plt.show()

print("Selesai! Visualisasi data kelembapan telah diperbarui.")
```
